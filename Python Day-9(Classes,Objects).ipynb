## Operator Overloading:

Screenshot 2025-09-21 202144.png

Operator overloading in Python allows the customization of how standard operators (like +, -, *, ==, etc.) behave when applied to user-defined objects. This feature enables objects of custom classes to interact with operators in a way that is intuitive and consistent with their intended behavior

```python
class PlayerClass:
  team_size = 11 # this will be available to all inside this glass, also called global attribute
  def __init__(self, fname,lname,year): # init is like a constructor, here we're defining init to load the attribute. Only applicable to this
    self.first_name = fname
    self.last_name = lname
    self.birth_year = year
    self.scores = []

  def get_scores(self, score):
    self.scores.append(score)

  def get_avg_score(self):
    return sum(self.scores)/len(self.scores)

  def get_age(self):
    return self.birth_year

  def __str__(self): # in build method of a class
    return f'{self.first_name} {self.last_name}'

  def __gt__(self, other): # greater than operator overloaded in this function
    self_avg_score = self.get_avg_score() # here self will be david
    other_avg_score = other.get_avg_score() # and other is virat
    return self_avg_score > other_avg_score

  def __eq__(self, other):
    self_year = self.birth_year
    other_year = other.birth_year
    return self_year == other_year



# instance of the class:
virat = PlayerClass('Virat','Kohli', 1988)
david = PlayerClass('David', 'Warner', 1980)
# print(virat.first_name)
# print(virat.last_name)
# print(virat.birth_year)
virat.get_scores(28)
virat.get_scores(45)
virat.get_scores(13)

david.get_scores(30)
david.get_scores(29)
david.get_scores(17)

print(virat) # this will call the '__str__()' method
print(virat.scores)
print(virat.get_avg_score())
print(virat.get_age())

print(david) # this will call the '__str__()' method
print(david.scores)
print(david.get_avg_score())
print(david.get_age())

#print(david > virat) # this will give an error: '>' not supported between instances of 'PlayerClass' and 'PlayerClass'. So we have to use operator overloading
# the behavior is not defined that's why it's giving the error
# but we can define the behavior for the custom class, here it's (__gt__) => __gt__(self, other): Overloads the > operator for greater than comparison.

# After defining function for operator overloaded, then:
print(david > virat)

print(david == virat)
```

## Inheritance:

Inheritance in Python is a fundamental concept in Object-Oriented Programming (OOP) that allows a class (the "child" or "derived" class) to inherit attributes and methods from another class (the "parent" or "base" class). This promotes code reusability and establishes a hierarchical relationship between classes

Key Concepts:
- Parent Class (Base Class/Superclass): The class from which other classes - inherit properties and behaviors.
- Child Class (Derived Class/Subclass): The class that inherits properties and behaviors from a parent class.

```python
class ParentClass():
  def __init__(self, name):
    self.name = name

  def  greet(self):
    print(f'hello from {self.name} (parent class)')

class ChildClass(ParentClass):
  def __init__(self, name, age):
    super().__init__(name)  # Call the parent class's __init__ method
    self.age = age

  def introduce(self):
        print(f"My name is {self.name} and I am {self.age} years old.")


# Create objects
parent_obj = ParentClass("Mom")
child_obj = ChildClass("Son", 10)

# Call methods
parent_obj.greet()
child_obj.greet() # Child class inherits and can use parent's methods
child_obj.introduce()
```
