## Extend() Function: It extends the list

```python
list1 = [1,2,3,4]
list1.extend([4,5,6])
list1
```

## append() function:

```python
list2 = [1,2,3,4]
list2.append([4,5,6])
list2
```

```python
# In append when you pass a list, then it appends the full list as a new entity
```

## Del - command: used to delete some attribute from a list

```python
list2 = [1,2,3,4]
del list2[2]
list2
```

```python
## pop() - Delete the object based on the index value
# Default --> Removes the last element
```

```python
list1 = [11,12,22,33]
list1.pop(2)
list1
```

```python
# remove() --> Remove operation works on values
list2 = [12,33,44,32]
list2.remove(33)
list2
```

## Sorting Function sort()-->: Sorting a list

```python
# Default it sort in ascending order
```

```python
A = [2,5,3,7]
A.sort()
A
```

```python
# To sort in descending order:
A.sort(reverse=True)
A
```

## Difference between sort and sorted

```python
A = ["Oranges", "Strawberry", "Mango"]
B = A.sort()
print(A)
print(B)
```

```python
# When we want to store the sorted value into another variable, then we need to use 'sorted()'
```

```python
A = ["Oranges", "Strawberry", "Mango"]
B = sorted(A)
print(A)
print(B)
```

```python
# sort does not create another variable but when we use sorted() then we can assign it into another variable
```
