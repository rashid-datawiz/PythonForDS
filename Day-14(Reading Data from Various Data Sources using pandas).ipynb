# Reading Data From Different Sources

## Read json object and convert it into dataframe

```python
import pandas as pd
from io import StringIO # StringIO module is used to handle string data as file-like object
data = '{"employee_name": "James", "email": "james@gmail.com", "job_profile": [{"title1":"Team Lead", "title2":"Sr. Developer"}]}'
df = pd.read_json(StringIO(data))
```

```python
df
```

## Convert Dataframe into json:

```python
df_j = df.to_json(orient='index')
df_j
```

```python
df_j = df.to_json(orient='records')
df_j
```

## Reading data from url:

```python
df = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data", header=None)
df
```

```python
## convert it back to CSV:
df.to_csv('wine.csv')
```

## Read Data from web html:

```python
url = "https://www.fdic.gov/resources/resolutions/bank-failures/failed-bank-list/"
df = pd.read_html(url)[0] # Take the fst element of the list
df
```

```python
url = "https://en.wikipedia.org/wiki/Mobile_country_code"
df = pd.read_html(url, match='Counrty', header=0)
df
```
