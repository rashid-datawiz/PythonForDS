## Function Examples:

```python
## Even or odd:

def evn_odd(num):
  # here num is argument
  if num%2==0:
    print(f"{num} is even")
  else:
    print(f"{num} is odd")

evn_odd(23)

```

```python
## Factorial:
# 4! = 4*3*2*1 = 24
```

```python
def fac_num(n):
  fact = 1

  for i in range(1,n+1):
    fact = fact*i
  return fact

fac_num(4)
```

```python
# Sum of all natural numbers till the number provided

def sum_numbers(n):
    num = 0

    for i in range(1,n+1):
        num = num+i
    return num


sum_numbers(3)


'''
n = 3
for i in range(1,4):
    num = 0 + 1 --> num = 1


for i = 2
    num = num + i
    num = 1 + 2 = 3 --> num = 3

for i = 3
    num = num + i
    num = 3+3 --> num = 6 --> Returned
'''
```

## Lambda function:

### Create a lambda function that adds 10 to the number passed as an arguement, and print the result

# lambda argument: expressions

```python
x = lambda a : a + 10
```

```python
print(x(5))
```

### Create a lambda function that checks a number is even or odd:

```python
x  = lambda a: 'even' if a%2 == 0 else 'odd'
```

```python
print(x(7))
```

### Create a lambda function that adds two integers


```python
x = lambda a,b : a + b
```

```python
print(x(5,15))
```
