# Sets
- Sets are a type of collection like lists, tuples storing mixed data
- Sets are enclosed within curly brackets and elements are written as comma-separated
- Sets are unordered
- Sets does not allow duplicates

```python
s = {2,3,4,4,5,5,2,1,7,8,8}
type(s)
```

```python
s
```

```python
len(s)
```

```python
# Typecasting sets:
list = [2,4,3,5,7,6,2,3]

new_s = set(list)
new_s
```

```python
type(new_s)
```

## Set Operations:

```python
A = {0,2,4,6,8}
B = {1,2,3,4,5}
```

```python
#union(): Combination of all elements or operations

print(A | B)
print(A.union(B))
print(len(A.union(B)))
```

```python
#intersection() : Only the common elements
print(A & B)
print(A.intersection(B))
```

```python
#difference: elements which is in A but not in B

print(A - B)
print(A.difference(B))
```

```python
#symmetric difference: The whole outer join minus the inner join

print(A ^ B)
```

Why do we use Sets?
Let’s take a look at an example to understand this better:

Let’s say you have a huge list that contains student grades:

Grades = ['A', 'A', 'B', 'C', 'D', 'B', 'B', 'C', 'D', 'E', 'C', 'C', 'A', 'B', 'F', 'D', 'C', 'B', 'C', 'A', 'B', 'F', 'B', 'A', 'E', 'B', 'B', 'C', 'D'...]

You want to identify distinct grades allotted to students. Obviously, you cannot check every element of this list; instead, we make use of sets which gets our job done here.

```python
Grades = ["A", "A", "B", "C", "D", "B", "B", "C", "D", "E", "C", "C", "A", "B", "F", "D", "C", "B", "C", "A", "B", "F", "B", "A", "E", "B", "B", "C", "D"]

set(Grades)
```

# Dictionaries:

## Dictionary:
- Dictionary stores element as keys and value pairs
- The key is like an index, its always unique &      immutable
- The values are the objects that contain information
- Values are accessed using their keys
- Each key is followed by a value separated by a colon
- The values can be immutable, mutable, & duplicates
- Each key and value pair is separated by a comma enclosed inside curly brackets

```python
# Creating a dictionary

d1 = {"bangladesh": "BDT", "USA": "USD", "Hong Kong": "HKD"}
# Accessing value using keys

d1["bangladesh"]
```

```python
type(d1)
```

```python
len(d1)
```

```python
# Replacing the value for a key in a dictionary

d1["bangladesh"] = "taka"
d1
```

```python
# creating new entity:
d1['New'] = 'N'
d1
```

```python
# Deleting a key value pair:

del d1["New"]

d1
```

```python
# Sorting a dictionary

sorted(d1)
```

```python
# values() method

d1.values()
```

```python
# keys() method

d1.keys()
```

```python
# update() method

d1.update({"bangladesh": "BDT"})
d1["bangladesh"] = "BDT"

d1
```
