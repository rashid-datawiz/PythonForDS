## Data MAnipulation with Dataframe:

```python
df
```

```python
## Add a new column:
df['Salary'] = [50000, 60000, 70000]
df
```

```python
# Remove a column:

# axis parameter values
# axis=0 or axis='index': Drops labels from the rows.
# axis=1 or axis='columns': Drops labels from the columns

df.drop('Salary', axis=1, inplace=True) # inplace=True means change to the original dataset
df
```

```python
## Add Age to the column:

df['Age'] = df['Age']+1
df
```
