# Variables:

```python
# Variables are storage for values
# Ex-
a = 5
b = 'apple'
print(a,b)
```

```python
# Variables can store one item at a time,
# If we assign new value to the previous variable, then the previous value is overwritten

a = 5
b = 10
print(a,b)
a = b
print(a)
```

```python
# we can also assign multiple variables at the same line

a,b = 10, 20
print(a,b)
```

# Operators:

## Arithmetic Operators:

```python
a = 5
b = 7
print(a+b) # Add
print(a-b) # Sub
print(a*b) # Multi
print(a/b) # Div
```

```python
# exponent
print(a ** b)
```

```python
# Python operations starts from Left to Right:

16- 2 + 4 * 1 / 5
```

```python
16- (2 + (4 * (1 / 5)))
```

## Comparison Operators:

```python
# Equal: == (also boolean operator)
5==6
```

```python
# Not Equal !=
5 != 6
```

## PYTHON ASSIGNMENT OPERATORS:
Operator	      Example	Same As
--------   |     ------- ---------
=	              x = 5	  x = 5
+=	            x += 3	x = x + 3
-=	            x -= 3	x = x - 3
*=	            x *= 3	x = x * 3
/=	            x /= 3	x = x / 3
%=	            x %= 3	x = x % 3
//=	            x //= 3	x = x // 3
**=	            x **= 3	x = x ** 3
&=	            x &= 3	x = x & 3
|=	            x |= 3	x = x | 3
^=	            x ^= 3	x = x ^ 3
>>=	            x >>= 3	x = x >> 3
<<=	            x <<= 3	x = x << 3
:=	            print(x := 3)	x = 3 print(x)
