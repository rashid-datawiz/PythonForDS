# Strings:

print("Newton's 3rd law states: "For every action there's an equal but opposite reaction"")

# It's giving syntax error! So how to solve it?

```python
# Solution: (Escape character:-> '\')
print('Newton\'s 3rd law states: "For every action there\'s an equal but opposite reaction"')
```
